```python
import os
import shutil
import pandas as pd

ROOT = "/content/Grocery-Dataset"
OUTPUT = "/content/fridge_data_cls"

# 1. טעינת הקלאסים עם טיפול ברווחים וסימני הפרדה
classes_csv_path = os.path.join(ROOT, 'dataset', 'classes.csv')

# נקרא את הקובץ בלי להניח שהכל פסיקים, וננקה רווחים משמות העמודות
classes_df = pd.read_csv(classes_csv_path, skipinitialspace=True)
classes_df.columns = classes_df.columns.str.strip() # ניקוי רווחים משמות העמודות

# נבדוק מה השמות האמיתיים של העמודות כדי שלא ניפול שוב
print("Columns found in CSV:", list(classes_df.columns))

# יצירת המפה (מוודא שאנחנו משתמשים בשמות הנכונים גם אם הם עם רווחים)
id_col = 'Class ID' if 'Class ID' in classes_df.columns else classes_df.columns[1]
name_col = 'Class Name' if 'Class Name' in classes_df.columns else classes_df.columns[0]

class_map = dict(zip(classes_df[id_col], classes_df[name_col]))

def prepare_colab_data(txt_file, subset_name):
    txt_path = os.path.join(ROOT, 'dataset', txt_file)
    if not os.path.exists(txt_path):
        print(f"Warning: {txt_file} not found!")
        return

    with open(txt_path, 'r') as f:
        for line in f:
            # ניקוי השורה והפרדה (מטפל בפסיקים ורווחים)
            parts = line.replace(',', ' ').strip().split()
            if len(parts) < 2: continue

            img_rel_path = parts[0]
            try:
                class_id = int(parts[1])
                class_name = class_map[class_id].replace("-", "_").replace(" ", "_")

                src = os.path.join(ROOT, 'dataset', img_rel_path)
                dst_dir = os.path.join(OUTPUT, subset_name, class_name)

                os.makedirs(dst_dir, exist_ok=True)
                if os.path.exists(src):
                    shutil.copy(src, dst_dir)
            except (ValueError, KeyError):
                continue

print("מארגן תיקיות מחדש...")
if os.path.exists(OUTPUT): shutil.rmtree(OUTPUT) # ניקוי ניסיונות קודמים
prepare_colab_data('train.txt', 'train')
prepare_colab_data('val.txt', 'val')
print("הסתיים בהצלחה! עכשיו אפשר להמשיך לשלב 4 (האימון).")
```

```python
from ultralytics import YOLO
from google.colab import files
import cv2
from google.colab.patches import cv2_imshow

# 1. טעינת המודל שאימנו
model_path = '/content/runs/classify/fridge_classification/weights/best.pt'
model = YOLO(model_path)

# 2. העלאת תמונה מהמחשב
print("Upload an image of a grocery item:")
uploaded = files.upload()

for filename in uploaded.keys():
    # 3. ביצוע חיזוי
    results = model.predict(filename)

    # 4. חילוץ התוצאה הכי גבוהה
    names_dict = results[0].names
    probs = results[0].probs
    top1_idx = probs.top1
    top1_conf = probs.top1conf.item()
    prediction = names_dict[top1_idx]

    # 5. הצגת התמונה עם הכיתוב
    img = cv2.imread(filename)
    print(f"\n--- Result ---")
    print(f"Prediction: {prediction}")
    print(f"Confidence: {top1_conf:.2f}")

    # הוספת טקסט על התמונה
    cv2.putText(img, f"{prediction} ({top1_conf:.2f})", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2_imshow(img)
```

```python
import torch
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

# 1. טעינת המודל (נשתמש במודל ה-Detection שאימנו)
model = YOLO('/content/runs/detect/Fridge_9000_Final/weights/best.pt')

# 2. העלאת התמונה שלכם
print("העלו תמונה של חלב או מוצר אחר:")
uploaded = files.upload()

for filename in uploaded.keys():
    # 3. הרצה עם סף נמוך מאוד (0.01) כדי לראות "ניצוצות" של זיהוי
    results = model.predict(source=filename, conf=0.01, imgsz=640)

    # 4. הדפסת רשימת כל הדברים שהוא חושב שהוא רואה
    print(f"\n--- ניתוח גולמי של המודל ---")
    boxes = results[0].boxes
    if len(boxes) == 0:
        print("המודל לא מזהה שום אובייקט, אפילו בביטחון נמוך.")
    else:
        for box in boxes:
            cls_id = int(box.cls[0])
            conf = float(box.conf[0])
            name = model.names[cls_id]
            print(f"מוצר: {name} | ביטחון: {conf:.4f}")

    # 5. הצגת התמונה עם כל הריבועים (גם אלו בביטחון נמוך)
    res_plotted = results[0].plot()
    cv2_imshow(res_plotted)
```
